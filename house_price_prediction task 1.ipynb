```python

# House Price Prediction using Linear Regression

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error

# Step 1: Load datasets
train = pd.read_csv(r"D:\train_sales.csv")
test = pd.read_csv(r"D:\test.csv")

# Step 2: Select relevant features
features = ['GrLivArea', 'BedroomAbvGr', 'FullBath', 'HalfBath']
target = 'SalePrice'

X_train = train[features].fillna(0)
y_train = train[target].fillna(0)
X_test = test[features].fillna(0)

# Step 3: Train linear regression model
model = LinearRegression()
model.fit(X_train, y_train)

# Step 4: Make predictions
y_pred = model.predict(X_test)

# Step 5: Evaluate on training data
train_pred = model.predict(X_train)
rmse = np.sqrt(mean_squared_error(y_train, train_pred))
print("Training RMSE:", rmse)

# Step 6: Save results in sample submission format
submission = pd.read_csv(r"D:\sample_submission.csv")
submission['SalePrice'] = y_pred
submission.to_csv("my_submission.csv", index=False)
print("Prediction file saved as my_submission.csv")

# Step 7: Visualize feature relationships
for col in features:
    plt.scatter(train[col], y_train, alpha=0.5)
    plt.xlabel(col)
    plt.ylabel("SalePrice")
    plt.title(f"{col} vs SalePrice")
    plt.show()
```
